*Prediksi Harga Rumah Menggunakan Neural Network*

```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('drive/MyDrive/Machine Learning/data_house_price.csv')
```

```python
dataset
```

```python
bedrooms = dataset["bedrooms"]
bathrooms = dataset["bathrooms"]
luas = dataset['sqft_living']
floor = dataset['floors']
prices = dataset['price']
```

```python
house_all = []

for index in range(len(bedrooms)):
  house = []
  house.append(bedrooms[index])
  house.append(bathrooms[index])
  house.append(luas[index])
  house.append(floor[index])

  house_all.append(house)
```

```python
print(house_all)
house_all = np.array(house_all)
prices = np.array(prices)
```

```python
model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(units = 1, input_shape = [4])
])
```

```python
model.compile(
    loss='mae',
    optimizer='sgd',
    metrics=['accuracy']
)
```

```python
model.fit(
    house_all, prices, epochs=100
)
```

```python
kamar_tidur = int(input("Masukkan jumlah kamar tidur: "))
kamar_mandi = int(input("Berapa kamar mandi : "))
luas_rumah = int(input("luas rumah yang diinginkan : "))
jumlah_lantai = int(input("Lantai berapa : "))
```

```python
prediksi_harga = model.predict([[kamar_tidur, kamar_mandi, luas_rumah, jumlah_lantai]])
print("Rumah impian anda seharga {} USD".format(prediksi_harga))
```
